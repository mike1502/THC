# ¿Qué es Matplotlib?
[regresar al contenido](#c1)

Matplotlib es una biblioteca de trazado de gráficos de bajo nivel en Python que sirve como una utilidad de visualización.

Matplotlib fue creado por John D. Hunter.

Matplotlib es de código abierto y podemos usarlo libremente.

Matplotlib está principalmente escrito en Python, con algunas partes escritas en C, Objective-C y Javascript para garantizar la compatibilidad con diferentes plataformas.

## Dibujar puntos en coordenadas x e y
[regresar al contenido](#c1)

La función `plot()` se utiliza para dibujar puntos (marcadores) en un diagrama.

De forma predeterminada, la función `plot()` dibuja una línea de punto a punto.

La función toma parámetros para especificar los puntos en el diagrama.

El parámetro 1 es un array que contiene los puntos en el eje x.

El parámetro 2 es un array que contiene los puntos en el eje y.

Si necesitamos trazar una línea desde (1, 3) hasta (8, 10), debemos pasar dos arrays [1, 8] y [3, 10] a la función plot.

```python
import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([1, 8])
ypoints = np.array([3, 10])

plt.plot(xpoints, ypoints)
plt.show()
```

<a name='2-2'></a>

## Dibujar puntos sin linea
[regresar al contenido](#c1)

Para trazar solo los marcadores en un gráfico, puedes utilizar la notación de cadena abreviada `'o'`, que representa 'anillos'. Esto significa que solo se mostrarán los marcadores en lugar de una línea que los conecte

```python
import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([1, 8])
ypoints = np.array([3, 10])

plt.plot(xpoints, ypoints, 'o')
plt.show()
```

[regresar al contenido](#c1)

# Marcadores
[regresar al contenido](#c1)

Puedes utilizar el argumento de palabra clave "marker" para resaltar cada punto con un marcador específico. Esto te permite personalizar la apariencia de los puntos en el gráfico, utilizando diferentes marcadores como círculos, cruces, triángulos, etc.

```python
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = 'o')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = '*')
plt.show()
```

## Formato Strings fmt
[regresar al contenido](#c1)

También puedes utilizar la notación de cadena abreviada para especificar el marcador. Este parámetro también se llama "fmt" y se escribe con la siguiente sintaxis:

marcador|línea|color

```python
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, 'o:r')
plt.show()
```

Con Pyplot, puedes utilizar las funciones `xlabel()` y `ylabel()` para establecer una etiqueta para los ejes x e y en un gráfico. Estas funciones te permiten agregar descripciones claras a tus gráficos para que los datos sean más comprensibles.

# Grid
[regresar al contenido](#c1)

Con Pyplot, puedes utilizar la función `grid()` para agregar líneas de cuadrícula a tu gráfico. Las líneas de cuadrícula son útiles para ayudar a visualizar y medir los valores en el gráfico, lo que facilita la interpretación de los datos representados.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.plot(x, y)

plt.grid()

plt.show()
```

Puedes utilizar el parámetro `"axis"` en la función `grid()` para especificar qué líneas de cuadrícula mostrar. Los valores válidos son: 'x', 'y' y 'both' (ambos). El valor predeterminado es 'both' (ambos), lo que significa que se mostrarán líneas de cuadrícula tanto en el eje x como en el eje y. Sin embargo, puedes personalizarlo según tus necesidades para mostrar solo las líneas de cuadrícula que desees.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.plot(x, y)

plt.grid(axis = 'x')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.plot(x, y)

plt.grid(axis = 'y')

plt.show()
```

# Propiedades del grid
[regresar al contenido](#c1)

También puedes establecer las propiedades de línea de la cuadrícula utilizando la función `grid()`, de la siguiente manera: `grid(color='color', linestyle='linestyle', linewidth=number)`. Esto te permite personalizar el color, el estilo de línea y el grosor de las líneas de la cuadrícula en tu gráfico. Puedes reemplazar 'color' con el nombre del color deseado (por ejemplo, 'blue' para azul), 'linestyle' con el estilo de línea deseado (por ejemplo, 'dotted' para punteado) y 'number' con el grosor de la línea deseado (por ejemplo, 1.5 para un grosor de línea de 1.5 puntos).

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.plot(x, y)

plt.grid(color = 'green', linestyle = '--', linewidth = 0.5)

plt.show()
```

# Subplots
[regresar al contenido](#c1)

```python
import matplotlib.pyplot as plt
import numpy as np

#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(1, 2, 1)
plt.plot(x,y)

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(1, 2, 2)
plt.plot(x,y)

plt.show()
```

La función `subplot()` toma tres argumentos que describen la disposición de la figura.

La disposición se organiza en filas y columnas, que están representadas por el primer y segundo argumento.

El tercer argumento representa el índice del gráfico actual.

```python
plt.subplot(1, 2, 1)
# La figura tiene 1 fila, 2 columnas y este gráfico es el primer gráfico.
```

```python
plt.subplot(1, 2, 2)
# La figura tiene 1 fila, 2 columnas y este gráfico es el segundo gráfico.
```

Dibuja 2 gráficos uno encima del otro:

```python
import matplotlib.pyplot as plt
import numpy as np

#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(2, 1, 1)
plt.plot(x,y)

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(2, 1, 2)
plt.plot(x,y)

plt.show()
```

Puedes dibujar tantos gráficos como desees en una figura, solo describe el número de filas, columnas y el índice del gráfico.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(2, 3, 1)
plt.plot(x,y)

x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(2, 3, 2)
plt.plot(x,y)

x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(2, 3, 3)
plt.plot(x,y)

x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(2, 3, 4)
plt.plot(x,y)

x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(2, 3, 5)
plt.plot(x,y)

x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(2, 3, 6)
plt.plot(x,y)

plt.show()
```

# Súper título
[regresar al contenido](#c1)

Puedes agregar el título de cada subfigura con el comando `suptitle()`

```python
import matplotlib.pyplot as plt
import numpy as np

#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(1, 2, 1)
plt.plot(x,y)
plt.title("Salidas")

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(1, 2, 2)
plt.plot(x,y)
plt.title("Entradas")

plt.suptitle("Mi tienda")
plt.show()
```
